**LINEAR REGRESSION - STUDENT MARKS**

**PREPROCESSING**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df=pd.read_csv('https://raw.githubusercontent.com/lambagayatri08/MachineLearning/main/Student_Marks.csv')
df.head()
```

```python
df.describe()
```

```python
for col in df.columns:
    if 'Marks' in col:
        del df[col]

df
```

```python
X1=df.head()
X1
```

```python
X1.plot.scatter('time_study','number_courses',color='red')
plt.show()
```

```python
for y in X1.columns:
  X1[y].plot.density()
  plt.show()
```

```python
X_train=df.drop('number_courses',axis=1)
y_train=df['number_courses']
```

**MODELLING**

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
model.coef_,model.intercept_
```

```python
X1.plot.scatter('time_study','number_courses',color='blue')
plt.plot(
    np.array(X1['time_study']),
    model.coef_[0]*np.array(X1['time_study'])+ model.intercept_,
    color='red'
)
plt.show()
```
